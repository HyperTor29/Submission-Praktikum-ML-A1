```python
# Import Library yang dibutuhkan
import pandas as pd
```

```python
# Memuat dataset dari url yang sudah ditentukan
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/hayes-roth/hayes-roth.data"

names = ['name', 'hobby', 'educational-level', 'marital-status', 'class']

dataset = pd.read_csv(url, names=names)
```

```python
# Melihat dimensi dataset
dataset.shape
```

```python
# Melihat isi dataset
dataset
```

```python
# Melakukan distribusi kelas data
dataset.groupby('class').size()
```

```python
# Melihat ringkasan statistik data
dataset.describe()
```

```python
# Melihat informasi dari data
dataset.info()
```

```python
# Import Library yang dibutuhkan
import matplotlib.pyplot as plt
```

```python
# Memvisualisasikan data menggunakan Plot Univariat
dataset.plot(kind='box', subplots=True,layout=(2,4), sharex=False, sharey=False)
plt.show()
```

Plot tersebut menggunakan boxplot untuk menunjukkan distribusi data dari masing-masing variabel. Boxplot adalah jenis plot yang menunjukkan distribusi data secara ringkas, yaitu :

1. Median (garis tengah)

2. Mean (rata-rata)

3. Q1 (kuartil pertama)

4. Q3 (kuartil ketiga)

5. Interquartile range (IQR)

6. Outlier (nilai yang berada di luar IQR)

Berdasarkan plot tersebut, dapat ditarik beberapa kesimpulan sebagai berikut:

1. Terdapat 3 kelas, yaitu kelas 1.0, kelas 2.0, dan kelas 3.0.

2. Kelas 3.0 memiliki jumlah siswa paling banyak, diikuti oleh kelas 2.0, dan kelas 1.0.

3. Hobi yang paling populer adalah 3, diikuti oleh 2, dan menonton 1.

4. Tingkat pendidikan yang paling umum adalah 3, diikuti oleh 2, dan 1.

5. Status perkawinan yang paling umum adalah 3, diikuti oleh 2, dan 1.

Berikut adalah penjelasan lebih rinci mengenai masing-masing variabel :

1. Name

Pada variabel name, boxplot menunjukkan bahwa jumlah siswa dengan nama yang berbeda-beda cukup bervariasi. Hal ini menunjukkan bahwa data berasal dari populasi yang heterogen.

2. Hobby

Pada variabel hobby, boxplot menunjukkan bahwa hobi yang paling populer adalah 3, diikuti oleh 2, dan menonton 1. Hal ini menunjukkan bahwa kebanyakan siswa memiliki hobi yang aktif dan produktif.

3. Educational Level

Pada variabel educational level, boxplot menunjukkan bahwa tingkat pendidikan yang paling umum adalah 3, diikuti oleh 2, dan 1. Hal ini menunjukkan bahwa kebanyakan siswa masih berusia muda dan belum menyelesaikan pendidikan tinggi.

4. Marital Status

Pada variabel marital status, boxplot menunjukkan bahwa status perkawinan yang paling umum adalah 3, diikuti oleh 2, dan 1. Hal ini menunjukkan bahwa kebanyakan siswa masih belum menikah.

5. Class

Pada variabel class, boxplot menunjukkan bahwa kelas 3.0 memiliki jumlah siswa paling banyak, diikuti oleh kelas 2.0, dan kelas 1.0. Hal ini menunjukkan bahwa kualitas siswa di kelas 3.0 lebih baik daripada kelas 2.0 dan kelas 1.0.

Secara keseluruhan, plot univariat tersebut memberikan gambaran umum mengenai karakteristik dari data yang dianalisis. Plot tersebut dapat digunakan untuk mengidentifikasi pola dan tren dalam data, serta untuk membandingkan karakteristik antar kelompok data.
